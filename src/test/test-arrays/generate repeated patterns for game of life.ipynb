```python
def copypattern(a, n, m,  y, x):
    new_pattern = [];
    for i in range(n):
        line = a[i*m:(i+1)*m];
        print(line)
        new_pattern = new_pattern + line*x;
    new_pattern = new_pattern*y;
    return new_pattern;

def print2darray(a, n, m):
    for i in range(n):
        for j in range(m):
            print(a[i*m + j],end=" ");
        print("");
    print("\n");

def lexformat(inputstr):
    if inputstr == 0: return "."
    if inputstr == 1: return "O"
    
def print2darray_lexformat(a, n, m):
    for i in range(n):
        for j in range(m):
            print(lexformat(a[i*m + j]),end=" ");
        print("");
    print("\n");    
    
# define your input array rows (n) and columns (m) and 

def add_border(innercells, n, m, bn, bm):
    # n = y, m = x
    """Given rows (n) and columns (m) it creates an array,
    that when interpreted as a 2D array has a border with 0s
    and all inner cells each have a randomly chosen 0 or 1."""
    an = n+bn*2
    am = m+bm*2
    a = [0]*(an*am);

    if len(innercells) != (n)*(m):
        print(f"the inner cells {len(innercells)} are not compatible with n={n} and m={m}, {m*n} ")
    
    for i in range(bn,an-bn):
        for j in range(bm,am-bm):
            a[i*am + j] = innercells[(i-bn)*m + (j-bm)]
    return a, an, am;

def write_cells_to_file(file_name, inputcells, n, m):
    file = open(file_name,"w")
    file.write(str(n)+" ")
    file.write(str(m)+" ")
    j = len(inputcells)
    for i in range(j):
        file.write(str(inputcells[i])+" ")
    file.close()

def translate(encoded_message,y,x,file_name,bx,by):
    encoded_message = encoded_message.replace('.', '0')
    encoded_message = encoded_message.replace('O', '1')
    decoded_message =[]
    for i in range(len(encoded_message)):
        decoded_message = decoded_message + [int(encoded_message[i])]
    decoded_array,y_new,x_new = add_border(decoded_message,y,x, by, bx)
    file_name_padded = "test_"+file_name+"_"+str(y_new)+"x"+str(x_new)+".in"
    print2darray_lexformat(decoded_array,y_new,x_new)
    write_cells_to_file(file_name_padded,decoded_array,y_new,x_new)
    #return decoded_array,y_new,x_new;
    return decoded_array
 
def pad_N_package(decoded_message,y,x,file_name,bx,by):
    decoded_array,y_new,x_new = add_border(decoded_message,y,x, by, bx)
    file_name_padded = "test_"+file_name+"_"+str(y_new)+"x"+str(x_new)+".in"
    print2darray_lexformat(decoded_array,y_new,x_new)
    write_cells_to_file(file_name_padded,decoded_array,y_new,x_new)
    return decoded_array
 
```
